# load libraries, set paths

```python
import numpy as np
import pandas as pd
from analysis_functions import *
from functools import lru_cache
import os

base_dir = os.getcwd() # get dir of current file (this notebook)
```

# define game constraints
basic rules, numbers, and structures of the minigame, collected by manually recording data from each round

```python
# possible user actions
actions = {
    "small": np.arange(1,21),
    "large": np.arange(20,46),
    "very_small": np.arange(2,10)
}

# round structures
round1 = [
    {"conv":None,"win":(70,100)},
    {"conv":(60,70),"win":(70,100)},
    {"conv":(60,70),"win":(70,100)},
    {"conv":(65,75),"win":(75,100)},
    {"conv":(65,75),"win":(75,100)},
    {"conv":(70,80),"win":(80,100)}
]

round2 = [
    {"conv":(60,70),"win":(70,100)},
    {"conv":(60,70),"win":(70,100)},
    {"conv":(65,75),"win":(75,100)},
    {"conv":(65,75),"win":(75,100)},
    {"conv":(70,80),"win":(80,100)},
    {"conv":(70,85),"win":(85,100)},
    {"conv":(75,85),"win":(85,100)},
    {"conv":(75,85),"win":(85,100)}
]

round3 = [
    {"conv":(65,75),"win":(75,100)},
    {"conv":(65,75),"win":(75,100)},
    {"conv":(70,80),"win":(80,100)},
    {"conv":(70,80),"win":(80,100)},
    {"conv":(75,85),"win":(85,100)},
    {"conv":(75,85),"win":(85,100)},
    {"conv":(80,90),"win":(90,100)},
    {"conv":(80,98),"win":(98,100)},
    {"conv":(90,95),"win":(95,100)},
    {"conv":(90,95),"win":(95,100)}
]
```

# define game assumptions
these could be empirically verified, but for simplicity, I make some initial assumptions

```python
# assume values for each range of actions are uniformly distributed
action_probs = {
    a: np.ones(len(v))/len(v)
    for a,v in actions.items()
}

# assume probability of succesful convincing is 0.5
p_convince = 0.5
```

# optimal table computation

```python
# compute separate policies based on each rounds' convice/win ranges.
policy_r1 = compute_policy(round1, actions, action_probs, p_convince)
policy_r2 = compute_policy(round2, actions, action_probs, p_convince)
policy_r3 = compute_policy(round3, actions, action_probs, p_convince)
```

# optimal table save/output

```python
# add round labels 
policy_r1.insert(0, "round", 1)
policy_r2.insert(0, "round", 2)
policy_r3.insert(0, "round", 3)

# define save path 
save_optimal_tbl_path = os.path.join(base_dir, "data", "optimal_table.csv")

# concatenate dfs as one .csv, saved to data folder
optimal_tbl_df = pd.concat([policy_r1, policy_r2, policy_r3], ignore_index=True)
optimal_tbl_df.to_csv(save_optimal_tbl_path, index=False)
```
